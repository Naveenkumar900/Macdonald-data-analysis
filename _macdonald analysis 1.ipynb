```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
category_scaled = scaler.fit_transform(category)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(category_scaled)
```

```python
labels = kmeans.predict(category_scaled)
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(category_scaled, labels)
print(score)

plt.scatter(category_scaled[:, 0], category_scaled[:, 1], c=labels)
plt.show()
```
